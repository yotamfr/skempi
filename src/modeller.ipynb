```python
for i in tqdm(range(len(skempi_df)), desc="rows processed"):
    row = skempi_df.loc[i]
    t = row.Protein.split("_")
    struct = SkempiStruct(*t)
    struct.compute_dist_mat()
    mutations = parse_mutations(row["Mutation(s)_cleaned"])
    record = SkempiRecord(struct, mutations, row.DDG, group=0, minus_ddg=False)
    rrrecord = reversed(record)
```
