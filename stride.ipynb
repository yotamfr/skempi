```python
import os
import sys
import subprocess

import numpy as np
import pandas as pd

from io import StringIO
```

```python
skempi_df = pd.read_csv(os.path.join('data', 'SKEMPI_1.1.csv'), sep=';')
skempi_df.head(4)
```

```python
lim = None
pdbs = set(skempi_df[:lim]["Protein"])
n = len(pdbs) 
for i, protein in enumerate(pdbs):
    pdb, chainA, chainB = protein.split('_')
    cline = "python2 stride.py %s %s %s" % (pdb, chainA, chainB)
    assert os.WEXITSTATUS(os.system(cline)) == 0
    sys.stdout.write("\r{0:.0f}%".format(100.0 * i/n))
```
